## reading the csv files

```python
#using python built in module named csv
#importing the csv module
import csv
rows = []
with open(r"C:\Users\Hp\haberman.csv","r") as csvFile:
    csvread = csv.reader(csvFile)
    
    #reading all data rows
    for i in csvread:
        rows.append(i)
    
print('\nFirst 10 rows are:\n')
for row in rows[:10]:
    for col in row:
        print("%10s"%col,end=" "),
    print('\n')
```
